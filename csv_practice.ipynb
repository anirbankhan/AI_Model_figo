```python
#use of converters for changing values of a column
```

```python
def rename(name):
  if name == 'Sri Lanka':
     return 'SL'
  else:
    return name
```

```python
rename('Sri Lanka')
```

```python
rename('India')
```

```python
pd.read_csv('customers-100.csv', converters={'Country': rename})
```
