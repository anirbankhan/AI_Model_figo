```python
# CCA - Complete case analysys
import pandas as pd
```

```python
df = pd.read_csv('data_science_job.csv')
```

```python
# to get the missing value in percenntage
df.isnull().mean()*100
```

```python
df.columns
```

```python
# will consider the columns  for CCA where missing values are less than 5 percent
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0 ]
```

```python
cols
```

```python
# dropping records for this columns if there is a missing values
new_df = df[cols].dropna()
```
