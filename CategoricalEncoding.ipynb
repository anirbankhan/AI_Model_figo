```python
# Categorical Data Encoding
```

```python
# categorical data is 2 types - nominal data ( eg - states , where We cant apply any ordering )
# ordinal data  -( like degresss - mp, hs, graduation, masters , where we can apply ordeing )
# oridanl data encoders is applied in X (input column ), label data encoders applied on Y ( output data)
# one hot encoder is applied on nominal data in input column , means where We cant apply any ordering
```

```python
import pandas as pd
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df2 = df.head(10)
```

```python
df3 = df2.iloc[:,2:5]
```

```python
df2
```

```python
df3
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df3.iloc[:,0:2],df3.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories =[["high school","some college","bachelor's degree","master's degree","associate's degree"],["free/reduced","standard"]])
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
# level encoder is strictly used for labelled data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
