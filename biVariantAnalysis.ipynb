```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# multi variant analysis for numerical/categorical data set
sns.barplot(x='pclass', y='fare', hue='sex',data=titanic)
```

```python
sns.barplot(x='pclass', y='age', hue='sex', data=titanic)
```

```python
# box plot for bi-variant analysys
sns.boxplot(x='sex', y='age', data=titanic)
```

```python
# box plot for multi-variant analysys
sns.boxplot(x='sex', y='age', hue='pclass', data=titanic)
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
```

```python
sns.distplot(titanic[titanic['survived'] ==0 ]['age'], hist=False)
sns.distplot(titanic[titanic['survived'] ==1 ]['age'], hist=False)
```

```python
titanic.groupby('sex')['survived'].mean()
```

```python
titanic.groupby('pclass')['survived'].mean()
```

```python
# pair plot to analysis the relations between different numerical column
iris = sns.load_dataset('iris')
iris.head()
sns.pairplot(iris, hue='species')
```
